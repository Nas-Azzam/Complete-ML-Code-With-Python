## EDA

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_boston
```

```python
ds = pd.read_csv('housing.csv')
ds1  = load_boston()
ds1_df = pd.DataFrame(ds1.data, columns=ds1.feature_names)
ds1_df['MEDV'] = ds1.target
print(ds1['DESCR'])
```

```python
print(ds1_df.info())
```

```python
ds1_df.describe()
```

```python
print(ds1_df.columns)
```

```python
col_study = ['ZN', 'INDUS','NOX', 'RM',]
sns.pairplot(ds1_df[col_study],height=2.8)
plt.show()
```

```python
col_study =['PTRATIO','LSTAT', 'MEDV']
sns.pairplot(ds1_df[col_study],height=2.8)
plt.show()
```

## Correlation Analysis and Feature Selection

```python
pd.options.display.float_format = '{:,.2f}'.format
ds1_df.corr()
```
