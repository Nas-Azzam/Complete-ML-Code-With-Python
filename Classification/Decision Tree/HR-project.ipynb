```python
# removing the unrequired columns
# Method 1
num_col = list(ds.describe().columns)
col_catagorical = list(set(ds.columns).difference(num_col))
removed_col = ['EmployeeCount', 'StandardHours',  'EmployeeNumber']
col_numerical = [e for e in col_catagorical if e not in removed_col]
```

```python
col_catagorical
```

```python
ds['Attrition'] = ds['Attrition'].map({'Yes':1, 'No':0})
ds.head(2)
```

```python
col_catagorical.remove('Attrition')
col_catagorical
```

```python
ds_clean  = pd.get_dummies(ds[col_catagorical])
ds_clean.head(2)
```

```python
X = pd.concat([ds[num_col], ds_clean], axis=1)
X.head()
```

```python
y = ds['Attrition']
y.head()
```

```python
X.info()
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
classifier_  = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
```

```python
accuracy_score_train = accuracy_score(y_train, classifier_.predict(X_train))
print('Accuracy score for training set: {:.2f}'.format(accuracy_score_train))
accuracy_score_test = accuracy_score(y_test, classifier_.predict(X_test))
print('Accuracy score for test set: {:.2f}'.format(accuracy_score_test))
```

```python
# Classification report and confusion matrix for training set
print(classification_report(y_train, classifier_.predict(X_train)))
print('\n')
print(confusion_matrix(y_train, classifier_.predict(X_train)))
```

```python
# Classification report and confusion matrix for test set

print(confusion_matrix(y_test, classifier_.predict(X_test)))
print('\n')
print(classification_report(y_test, classifier_.predict(X_test)))

# ROC CURVE
roc_auc_score(y_test, classifier_.predict_proba(X_test)[:,1])
 
```
