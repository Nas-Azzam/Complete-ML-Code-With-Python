### Polynomial Regression

```python
# importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
# sns.set_style('blac')
```

```python
np.random.seed(42)
sample_size = 100

X = np.linspace(0, 10, sample_size)
rng = np.random.rand(sample_size) * 100

Y  = X ** 3 + rng + 100

plt.figure(figsize=(7,5))
plt.scatter(X, Y)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X.reshape(-1,1), Y)
lr_pred = lr.predict(X.reshape(-1,1))

print('R2 score: ', r2_score(Y, lr_pred))
plt.figure(figsize=(7,5))
plt.scatter(X, Y)
plt.plot(X, lr_pred, color='red')
plt.show()
```

```python
# polynomial regression
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg =  PolynomialFeatures(degree=2)
X_poly = poly_reg.fit_transform(X.reshape(-1,1))

lin_reg = LinearRegression()
lin_reg.fit(X_poly, Y)
lin_reg_pred = lin_reg.predict(X_poly)

 
```

```python
print('R2 score: ', r2_score(Y, lin_reg_pred))
plt.figure(figsize=(7,5))
plt.scatter(X, Y)
plt.plot(X, lin_reg_pred, color='red')
plt.show()
```
