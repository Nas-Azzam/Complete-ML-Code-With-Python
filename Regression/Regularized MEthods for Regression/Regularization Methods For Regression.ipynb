## Ridge Ression

```python
# importing Librareies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style('whitegrid') 
```

```python
# LInear Reggression
from sklearn.linear_model import LinearRegression
```

```python
np.random.seed(50)
sample_size = 100
x = np.random.rand(sample_size)* 10
y = 2*x + 5 + np.random.randn(sample_size)

lr = LinearRegression()
lr.fit(x.reshape(-1,1),y)
model_pred = lr.predict(x.reshape(-1,1))

plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red')
plt.plot(x,model_pred,color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()
print('Coefficients: \n', lr.coef_)
```

```python
# Adding Some Outliners

idx  = x.argmax()
y[idx] =  150
idx = x.argmin()
y[idx] = -150


plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red')
plt.plot(x,model_pred,color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.show()
print('Coefficients: \n', lr.coef_)

```

```python
# Rigde Regression (L2 Regularization)

from sklearn.linear_model import Ridge
```

```python
ridg_mod  = Ridge(alpha=0.1,normalize=True)
ridg_mod.fit(x.reshape(-1,1),y)
ridg_pred = ridg_mod.predict(x.reshape(-1,1))

plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red')
plt.plot(x,ridg_pred,color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Ridge Regression')
plt.show()
print('Coefficients: \n', ridg_mod.coef_)
```

```python
# Lasso regression (L1 Regularization)
from sklearn.linear_model import Lasso
```

```python
lasso_mode = Lasso(alpha=0.1,normalize=True)
lasso_mode.fit(x.reshape(-1,1),y)
lasso_pred = lasso_mode.predict(x.reshape(-1,1))

plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red')
plt.plot(x,lasso_pred,color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Lasso Regression')
plt.show()
print('Coefficients: \n', lasso_mode.coef_)
```

```python
# Elastic Net 
from sklearn.linear_model import ElasticNet
```

```python
en_mod = ElasticNet(alpha=0.1,l1_ratio=0.5,normalize=True)
en_mod.fit(x.reshape(-1,1),y)
en_pred = en_mod.predict(x.reshape(-1,1))

plt.figure(figsize=(10,5))
plt.scatter(x,y,color='red')
plt.plot(x,en_pred,color='blue')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Elastic Net Regression')
plt.show()
print('Coefficients: \n', en_mod.coef_)
```
